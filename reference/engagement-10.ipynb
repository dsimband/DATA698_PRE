```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go
```

```python
df = pd.read_csv('/content/2022MM_CAT.csv')
df
```

```python
org_columns = ['MM.com & Content', 'CS', 'CSP', 'Distribution','IS', 'MMSD','DX','ETX']
org_q_columns = ['MM.com & Content', 'CS', 'CSP', 'Distribution','IS', 'MMSD']
report_columns = ['CS', 'CSP','IS', 'MMSD','MM & DIST','DX','ETX']
report_q_columns = ['CS', 'CSP','IS', 'MMSD','MM & DIST']



#report_columns = ['CS', 'CSP','IS', 'MMSD','MM & DIST']
#report_columns = ['CS','MM & DIST']
#report_columns = ['CSP','MM & DIST']
#report_columns = ['IS','MM & DIST']
#report_columns = ['MMSD','MM & DIST']
```

```python
org_df = df.iloc[:,:3]
org_df.columns = org_df.iloc[0]
org_df = org_df.drop(labels=0, axis=0)
org_df['Group'] = ['MM.com & Content', 'CS', 'CSP', 'Distribution','IS', 'MMSD','DX','ETX']
org_df[['DatalgroupID','ResponseCount']] = org_df[['DatalgroupID','ResponseCount']].astype(int)
org_df
```

```python
org_count = dict(zip(org_df.Group, org_df.ResponseCount))
org_count
```

```python
ee_df = df.copy()

ee_df['Group'] = ['Response','MM.com & Content', 'CS', 'CSP', 'Distribution','IS', 'MMSD','DX','ETX']
ee_df = ee_df.drop(columns=['Unnamed: 0','Unnamed: 1'])

ee_df = ee_df.set_index('Group')
ee_df = ee_df.transpose()

ee_df.reset_index(inplace=True)
ee_df = ee_df.rename(columns={'DatalgroupID_DatagroupName':'Key','index':'Question'})
ee_df = ee_df.drop(labels=0, axis=0)

ee_df[org_columns] = ee_df[org_columns].astype(int)

ee_df['MM & DIST'] = round((ee_df['MM.com & Content'] / 100 * org_count['MM.com & Content'] 
                      + ee_df['Distribution'] / 100 * org_count['Distribution'] ) \
                      / (org_count['MM.com & Content'] + org_count['Distribution']),4) * 100

ee_df.Question = ee_df.Question.str.rstrip('(.1)')
ee_df.Question = ee_df.Question.str.rstrip('(.2)')

ee_df
```

```python
ee_cat_df = ee_df.iloc[:33]
ee_cat_df
```

```python
long_cat_df = pd.melt(ee_cat_df, id_vars=['Question','Response'], value_vars=report_columns)
long_cat_df = pd.pivot(long_cat_df, index=['Question','Group'], columns='Response' , values='value')
long_cat_df = long_cat_df.reset_index()
long_cat_df.set_index(['Question','Group'])
```

```python
fig = px.scatter_ternary(long_cat_df, a="Favorable", b="Neutral", c="Unfavorable", 
                         hover_name="Question", color="Group", size="Favorable", size_max=10)
fig.show()
```

```python
fig = px.box(
          data_frame = long_cat_df,
          y = 'Favorable', notched=True,
          color = 'Group', points="all", hover_data=["Question"]
)
fig.show()
```

```python
fig = px.box(
          data_frame = long_cat_df,
          y = 'Unfavorable', notched=True,
          color = 'Group', points="all", hover_data=["Question"]
)
fig.show()
```

```python
sort_df = long_cat_df.groupby(['Group']).apply(lambda x: x.sort_values(['Favorable'],ascending = False)).reset_index(drop = True)
sort_df.groupby(['Group']).head(5)
```

```python
m_df = pd.melt(ee_cat_df, id_vars=['Question','Response'], value_vars=report_columns)
m_df
```

```python
fig = px.bar(m_df, y='Question', x='value',
              color='Response' , barmode="stack", 
              height=2000, facet_row="Group", text_auto=True
             )
fig.show()
```
